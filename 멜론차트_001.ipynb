```python
import pandas as pd
```

```python
df_top50 = pd.read_csv('./weekly_chart/2021_10_04__2021_10_10장르종합주간차트____01_50.txt', sep='|')
display(df_top50)
```

```python
df_top100 = pd.read_csv('./weekly_chart/2021_10_04__2021_10_10장르종합주간차트____50_100.txt', sep='|')
display(df_top100)
```

```python
df = pd.concat([df_top50, df_top100], axis=0)
display(df)
```

```python
df.tail(10)
```

```python
df = df.drop(columns=['album_img'])
display(df)
```

```python
df['score'] = 101 - df['rank']
df
```

```python
df_sum = df[['artist_name', 'score']].groupby('artist_name').sum()
```

```python
df_sum
```

```python
df_sum = df_sum.sort_values(by='score', ascending=False)
```

```python
df_sum.head(15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_sum = df_sum.reset_index()
display(df_sum)
```

```python
fig, ax = plt.subplots(figsize=(10, 15))
sns.barplot(data=df_sum, x='score', y='artist_name')
plt.show()
```

```python
# 한글이 안 나올 때는...
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'NanumGothic'
#plt.rcParams['font.family'] = 'AppleGothic'

fig, ax = plt.subplots(figsize=(10, 15))
sns.barplot(data=df_sum.head(20), x='score', y='artist_name')
plt.show()
```
